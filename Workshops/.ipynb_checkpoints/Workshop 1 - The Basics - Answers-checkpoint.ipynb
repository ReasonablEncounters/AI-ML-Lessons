```python
import numpy as np
import random

# DEFINE THE CLASS
class NumberGame:
    
    # INITIALIZE PRIVATE CLASS VARIABLES
    def __init__(self):
        
        self.guess_number = None
        self.guess_count = 1
        self.random_number = random.randint(1,10)
    
    # FUNCTION (Guessing Game)
    def GuessTheNumber(self, guess):
        
        self.guess_number = int(guess)
        
        # determination 
        if self.guess_number == self.random_number:
            
            result = print("\n====== CONGRATULATIONS, YOU GOT IT CORRECT! " + "Only took " + str(self.guess_count) + " attempt(s)! ======")

        elif self.guess_number > self.random_number:
            
            result = print("\n====== TOO HIGH! ======")
            self.guess_count += 1

        elif self.guess_number < self.random_number:
            
            result = print("\n====== TOO LOW! ======")
            self.guess_count += 1

        else:
            
            result = print("\n====== Uhm, that's not a thing! ======")

        return result
```

### Observe the Output of the new Class

```python
# prompt the user to guess a number
# store it in a variable called guess
guess = input("Can you guess my number between 1 and 10?!\n")

# instantiate the class for use
number_game = NumberGame()

# call the function 
# with the user's guess as input
number_game.GuessTheNumber(guess)

# output the correct answer
print("The correct number was " + str(number_game.random_number))
```

### Create 2 New Python Functions
- Within your newly created Python class, create 2 new functions that utilize NumPy's built in tools as follows:
    - take input for a temperature
        - calculate & return Fahrenheit from Celsius input [GetFahrenheit()]
        - calculate & return Celsius from Fahrenheit input [GetCelsius()]

```python
## IMPORT NUMPY
import numpy as np
## IMPORT PANDAS
import pandas as pd

# NumPy and Pandas were imported in an earlier cell, 
# therefore they are already imported as long as that cell was executed
# Jupyter executes in the order that you run each cell, but it is best to run it top-down to avoid confusion

# DEFINE THE CLASS
class Calculate:

    
    # INITIALIZE PRIVATE CLASS VARIABLES
    # put your calculation constants here
        
    def __init__(self):
        
        self.fahrenheit_input = None
        self.celsius_input = None
        self.result = None
    
    # FUNCTION 1 - Celsius to Fahrenheit
    def CalculateFahrenheit(self, input_temperature):
        
        self.fahrenheit_input = int(input_temperature)
        
        self.result = ((input_temperature - 32) * .5556)
        
        return self.result
    
    # FUNCTION 2 - Fahrenheit to Celsius    
    def CalculateCelsius(self, input_temperature):
         
        self.celsius_input = int(input_temperature)
        
        self.result = ((input_temperature * 1.8) + 32)    
        
        return self.result
```

```python
# DEMONSTRATE YOUR NEW FUNCTIONS OUTPUT HERE
temperature = int(input("What temperature would you like to convert?\n"))

temperature_choice = input("Would you like to convert the temperature to Celsius or Fahrenheit?")

temperature_conversion = Calculate()



if temperature_choice == "F" or temperature_choice == "f":
    
    temperature_conversion.CalculateFahrenheit(temperature)
    
elif temperature_choice == "C" or temperature_choice == "c":
    
    temperature_conversion.CalculateCelsius(temperature)
    
else:
    
    print("Um, that's not an option!")

print("The temperature conversion = " + str(temperature_conversion.result))

```
